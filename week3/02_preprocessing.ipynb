Our group decided to only use LotSizeSquareFeet so I dropped the other related columns. I also removed columns containing names, email addresses, or office information because these features are unlikely to be informative for predicting listing closing prices.

```python
# Identify the proportion and number of null values for each column

def create_info_df(df):
    df_info = pd.DataFrame()
    df_info['prop_null'] = df.isna().sum()/len(df)
    df_info['num_null'] = df.isna().sum()
    df_info['data_types'] = df.dtypes
    return df_info.sort_values(by='prop_null', ascending=False)

create_info_df(df)
```

Based on these proportions, I chose to drop columns with at least 95% missing values. Since fewer than 5% of the entries in these columns contain data, they are unlikely to contribute meaningful predictive information, and most values would effectively be treated as unknown. 

```python
# Drop columns with a proportion of 95% or more missing values

drop = ['MiddleOrJuniorSchoolDistrict', 'TaxAnnualAmount', 'CoveredSpaces',
       'BusinessType', 'ElementarySchoolDistrict', 'FireplacesTotal',
       'TaxYear', 'AboveGradeFinishedArea', 'WaterfrontYN',
       'BelowGradeFinishedArea', 'BasementYN', 'BuilderName']
df = df.drop(columns=drop)
df
```

#### Dealing with Missing Values
- Drop null rows in columns with less than 5 missing values
- Fill NAs in categorical columns with 'Unknown'
- Flag missingness in rows where imputation was used
- Use median imputation or a missingness flag for numerical-categorical columns to preserve datatypes
- Assume that a missing value in a boolean column means False and fill accordingly
- Use median imputation on numerical rows to avoid outlier skew
- Assume missing `AssociationFee`, `GarageSpaces`, and `ParkingTotal` mean 0 and fill accordingly
- Drop columns with 0 values in `ClosePrice`, `LivingArea`, `BathroomsTotalInteger`, and `DaysOnMarket` as a 0 value is highly improbable

```python
# Categorize columns

cat_col = ['Flooring', 'UnparsedAddress', 'AssociationFeeFrequency', 
        'MLSAreaMajor', 'ElementarySchool', 'SubdivisionName', 'City', 
        'PurchaseContractDate', 'MiddleOrJuniorSchool', 'HighSchool',
        'HighSchoolDistrict', 'PostalCode', 'Levels', 'ListingKey', 'CloseDate', 
        'PropertyType', 'ListingKeyNumeric', 'CountyOrParish', 'MlsStatus', 
        'PropertySubType', 'ListingId', 'ContractStatusChangeDate',
        'ListingContractDate', 'StateOrProvince' ]

bool_col = ['ViewYN', 'PoolPrivateYN', 'AttachedGarageYN', 'FireplaceYN', 'NewConstructionYN']

num_col = ['LivingArea', 'BuildingAreaTotal', 'LotSizeSquareFeet', 'ListPrice', 
            'BathroomsTotalInteger', 'Stories', 'MainLevelBedrooms', 'BedroomsTotal', 'DaysOnMarket']
```

```python
# Drop null rows in columns with a very small number of missing values 
df = df.dropna(subset = ['PostalCode', 'ParkingTotal', 'Longitude', 'Latitude'])

# Fill missing categorical values with 'Unknown'
df[cat_col] = df[cat_col].fillna('Unknown')

# Flag missingness in columns where imputation is used
for col in df.columns:
    if df[col].isna().sum() > 0:
        df[f'{col}_was_missing'] = df[col].isna().astype(int) # 1 for missing, 0 for present

# Handle numerical-categorical columns with median or a specific numeric flag (like -1)
df['YearBuilt'] = df['YearBuilt'].fillna(df['YearBuilt'].median())
df['Latitude'] = df['Latitude'].fillna(df['Latitude'].median())
df['Longitude'] = df['Longitude'].fillna(df['Longitude'].median())
df['StreetNumberNumeric'] = df['StreetNumberNumeric'].fillna(-1)

# Convert to boolean type and fill NaNs with False
df[bool_col] = df[bool_col].fillna(False).astype(bool)

# Impute with the median value of the column
for col in num_col:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

# Special case: AssociationFee
# If an association fee is missing, it usually means there is no HOA fee (0.0)
if 'AssociationFee' in df.columns:
    df['AssociationFee'] = df['AssociationFee'].fillna(0.0)

# Fill missing parking/garage counts with 0
for col in ['GarageSpaces', 'ParkingTotal']:
    if col in df.columns:
        df[col] = df[col].fillna(0)

df = df[(df["ClosePrice"]>0) & (df["LivingArea"]>0) & (df["BathroomsTotalInteger"]>0) & (df["DaysOnMarket"]>0)]
df.head()
```

#### Encoding Columns
- Convert boolean columns to int values of 0 and 1
- Columns `Flooring` and `Levels` contained lists of values, use MultiLabelBinarizer to encode them
- Map ordinal values of `AssociationFeeFrequency` to their corresponding numerical values
- One hot encode a set of critical categorical columns instead of all to reduce column amount


```python
# Encode Boolean columns
encode_df = df.copy()
encode_df[bool_col] = encode_df[bool_col].astype(int)

# Encode 'Flooring' and 'Levels' columns using MultiLabelBinarizer
from sklearn.preprocessing import MultiLabelBinarizer

cols = ['Flooring', 'Levels']
for col in cols:
    encode_df[col] = encode_df[col].str.split(',')
    mlb = MultiLabelBinarizer()
    encoded = pd.DataFrame(mlb.fit_transform(df[col]), columns=[f'{col}_{c}' for c in mlb.classes_], index=df.index)
    encode_df = encode_df.join(encoded)
    encode_df = encode_df.drop(columns = col)

# Map the ordinal values in 'AssociationFeeFrequency' to their associated numerical values
mapping = {
    'Unknown': 0,
    'Monthly': 1,
    'Quarterly': 2,
    'SemiAnnually': 3,
    'Annually': 4
}
encode_df['AssociationFeeFrequency'] = encode_df['AssociationFeeFrequency'].map(mapping)

# One hot encode subset of critical categorical columns
one_hot = ['CountyOrParish', 'StateOrProvince', 'City', 'PropertyType', 'PropertySubType', 'MlsStatus']
encode_df = pd.get_dummies(encode_df, columns=[col for col in one_hot if col in encode_df.columns], drop_first=True)

df = encode_df
df.head()
```

#### Standardizing Numerical Values

Looking at our continuous numerical features, the scales vary drastically from one column to the next. Leaving the data with these massive differences in ranges can cause the final model to not make the most accurate predictions. To fix this and boost our overall model performance, I standardized the continuous variables so they're all evaluated on the same scale.

```python
# Standardize continuous numerical columns

from sklearn.preprocessing import StandardScaler

scale = [
    "LivingArea",
    "LotSizeSquareFeet",
    "AssociationFee",
    "DaysOnMarket"
]

scaler = StandardScaler()
df[scale] = scaler.fit_transform(df[scale])
df.head()
```
